```python
import pandas as pd
import os

if os.path.exists("Lms.csv") == False:
    df = pd.DataFrame({
        "User id": [],
        "Name": [],
        "Class": [],
        "Subject": []
    })
    df.to_csv("Lms.csv", index=False)

while True:

    print("LMS SYSTEM MENU")

    choice = int(input(" 1: Register | 2: Fetch | 3: Update | 4: Exit\nSelect: "))

    if choice == 1:

        df=pd.read_csv("Lms.csv")

        user_id = int(input("Enter User Id: "))
        if user_id in list(df["User id"]):
            print("ID already exists!")
            continue



        username = input("Enter your Name: ")
        if username in list(df["Name"]):
            print("Name already Exit")
            continue





        user_class = input("Enter Your Class: ")
        subject = input("Enter Subjects Name: ")


        id_list = list(df["User id"])
        name_list = list(df["Name"])
        class_list = list(df["Class"])
        subject_list = list(df["Subject"])


        id_list.append(user_id)
        name_list.append(username)
        class_list.append(user_class)
        subject_list.append(subject)


        studic = {
            "User id": id_list,
            "Name": name_list,
            "Class": class_list,
            "Subject": subject_list
        }

        df = pd.DataFrame(studic)
        df.to_csv("Lms.csv", index=False)

        print(f"\nCongratulations {username.capitalize()}! Registered successfully.")

    elif choice == 2:
        df=pd.read_csv("Lms.csv")


        search_id = int(input("Enter the ID to fetch: "))
        if search_id not in list(df["User id"]):
            print("Error: Account does not exist.")
            continue

        for i in range(len(df)):
            if df["User id"][i] == search_id:
                print(f"\n--- Student Found ---")
                print(f"ID: {df['User id'][i]}")
                print(f"Name: {df['Name'][i]}")
                print(f"Class: {df['Class'][i]}")
                print(f"Subject: {df['Subject'][i]}")




    elif choice == 3:
          df=pd.read_csv("Lms.csv")

          search_id = int(input("Enter the ID to fetch: "))
          if search_id not in list(df["User id"]):
             print("Error: Account does not exist.")
             continue

          while True:
              for i in range(len(df)):
                  if df["User id"][i]==search_id:
                        userchoice = input("What would you like to update? ")
                        print("Enter 1: ID | Enter 2: Name | Enter 3: Class | Enter 4: Subject")

                        if userchoice == '1':
                            df.loc[i, "User id"] = int(input("Enter new ID: "))
                            print("Update successful!")
                        elif userchoice == '2':
                            df.loc[i, "Name"] = input("Enter new Name: ")
                            print("Update successful!")
                        elif userchoice == '3':
                            df.loc[i, "Class"] = input("Enter new Class: ")
                            print("Update successful!")
                        elif userchoice == '4':
                            df.loc[i, "Subject"] = input("Enter new Subject: ")
                            print("Update successful!")
                        else:
                            print("Invalid Choice. Please try again.")
                            continue
                        break
              break



          df.to_csv("Lms.csv", index=False)



    elif choice == 4:
        print("Exiting system. Goodbye!")
        break
    else:
        print("Invalid Choice. Please try again.")
```
