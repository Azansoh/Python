```python
import pandas as pd
import random as r
import os



if os.path.exists("User_quiz.csv"):
    df = pd.read_csv("User_quiz.csv")
    user = df.to_dict(orient="list")
else:
    user={"User ID": [],
          "Username": [],
          "score":[],
          "Performance":[]
    }



def savedata2():
    df = pd.DataFrame(user)
    df.to_csv("User_quiz.csv", index=False)
    print("Data Saved Successfully in to User file !")





usedids = set(user["User ID"])

def uniqueid():
    while True:
        newid = r.randint(100, 999)
        if newid not in usedids:
            usedids.add(newid)
            return newid


while True:
    df=pd.read_csv("QuestionareDummy.csv")

    uid=uniqueid()
    name=input("Enter your full name: ").strip()

    if name in user["Username"]:
            print(f"Sorry! '{name}' you have already give Quiz!")
            continue




    randomnumberlist=[]
    for i in range(10):
        while True:
            randomnumber=r.randint(0,len(df)-1)
            if randomnumber not in randomnumberlist:
                randomnumberlist.append(randomnumber)
                break
            else:
                continue


    print("The Random Numbers is ",randomnumberlist)





    choicemap={1:"A",
           2:"B",
           3:"C",
           4:"D"
           }

    score=0
    for i in randomnumberlist:
        print(f"\nQuestion : {df['Question'][i]}")
        print(f"1. {df['Option A'][i]}")
        print(f"2. {df['Option B'][i]}")
        print(f"3. {df['Option C'][i]}")
        print(f"4. {df['Option D'][i]}")

        choice=int(input("Enter Your Correct Option (1,2,3,4) : "))

        if choice not in choicemap:
            print("Invalid option!")
            continue

        selectinp=choicemap.get(choice)

        if selectinp==df['Right Answer'][i]:
            score+=1
            print("Correct")

        else:
            print("Wrong Answer!")





    total_questions = len(randomnumberlist)
    percentage = (score / total_questions) * 100

    print(f"Your Total Score: {score} out of {total_questions}")
    print(f"Percentage: {percentage:.1f}%")

    if percentage == 100:
        remark="You're a genius!"
    elif percentage >= 70:
        remark="Great job!"
    else:
        remark="Keep practicing!"






    user["User ID"].append(uid)
    user["Username"].append(name)
    user["score"].append(score)
    user["Performance"].append(remark)






    savedata2()


    userchoi=input("Do you want to continue (y/n) :")
    if userchoi=="y" or userchoi=="Y":
        continue
    else:
        break





# yinn.c
# filhaner
# ballogy






























```
